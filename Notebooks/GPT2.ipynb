# Import and install dependancies

```
from google.colab import files
from google.colab import drive
%tensorflow_version 1.x
!pip install -q gpt-2-simple
import gpt_2_simple as gpt2
```

# Set working environment

```
# Mount Drive (Google) to store dataset and re-trained model
drive.mount('/content/drive')
```

```
# Set parameters for creating gpt-2 model specialised for a specific genre
GENRE = "western" # Typical scenario specification of the model (based on dataset available: "Western, Fantasy & Romance").
MODEL = "124M" # Model size desired.
```

# Download GPT-2 original model

```
gpt2.download_gpt2(model_name=MODEL)
```

```
# Get dataset from drive main folder (/content/drive)
file_name = GENRE + ".txt"
```

## Finetune GPT-2: Specilised GPT-2 model with dataset previously choosen

```
# Create tensorflow session
sess = gpt2.start_tf_sess()

# Train our gpt-2 model over 10000 epoch with our dataset
gpt2.finetune(sess,
              dataset=file_name,
              model_name=MODEL,
              steps=10000,
              restore_from='fresh',
              run_name='run1-' + GENRE + '-' + MODEL,
              print_every=10,
              sample_every=200,
              save_every=200
              )
```

# Generate Text From or GPT-2 trained model

```
# Generation of new scenario 
gpt2.generate(sess, run_name='run1-' + GENRE + '-' + MODEL)
```

```
# Generation of new scenario with only 100 words
gpt2.generate(sess, run_name='run1-' + GENRE + '-' + MODEL, length=100)
```

# Save artifact model in our Google drive

```
save = gpt2.copy_checkpoint_to_gdrive(run_name='run1-' + GENRE + '-' + MODEL)
```
